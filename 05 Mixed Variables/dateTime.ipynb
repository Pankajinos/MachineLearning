```python
date.drop(columns=['product_id','city_id','orders'], inplace=True)
```

```python
date['date'].dtype
```

it is a pandas object so we need to convert it into pandas date object

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date['year'] = date['date'].dt.year
date['month'] = date['date'].dt.month
date['month_name'] = date['date'].dt.month_name()
date['day'] = date['date'].dt.day
date['day_of_week'] = date['date'].dt.dayofweek
date['day_of_week_name'] = date['date'].dt.day_name()

date['quarter'] = date['date'].dt.quarter
date['week'] = date['date'].dt.isocalendar().week
```

```python
#is weekend
date['weekend'] = np.where(date['date'].dt.weekday.isin([5,6]), 1, 0)
```

```python
date['sem'] = np.where(date['date'].dt.quarter.isin([1, 2]), 1, 2)
```

```python
date.head()
```

```python
import datetime
today = datetime.datetime.today()
```

```python
today
```

```python
# date['year_elapsed'] = np.ceil((today-date['date']).dt.days/365)
date['seconds_elapsed'] = (today-date['date'])/np.timedelta64(1,'s')
date['hrs_elapsed'] = (today-date['date'])/np.timedelta64(1,'h')
date['month_elapsed'] = (today-date['date'])/np.timedelta64(1,'m')
```
