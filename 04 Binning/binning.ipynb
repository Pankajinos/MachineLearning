### Binning
- Bins continuous data into intervals
- There are three methods to do so
- strategy:
    1. uniform - All bins in each feature have identical widths.
    2. quatile - All bins in each feature have the same number of points.
    3. kmeans - Values in each bin have the same nearest center of a 1D k-means cluster.
- encoding
    1. ordinal - bins labeled with ordinal values
    2. onehot - encode the bins into a sparese matrix

```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import Binarizer

data = pd.DataFrame({
    'Age': [5, 5, 18, 25, 30, 40, 50, 50, 50]
})

kbins = KBinsDiscretizer(strategy='quantile', encode='ordinal', n_bins=5)
trans_data = kbins.fit_transform(data)
print(trans_data)
```

### Binarization 
- Binarization converts numerical features into binary values (0 or 1) using a threshold

```python
binarizer = Binarizer(threshold=30)
binary_data = binarizer.fit_transform(data)
binary_data
```
