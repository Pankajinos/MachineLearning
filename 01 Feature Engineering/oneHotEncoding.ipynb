For encoding nominal data we use oneHotEncoding
OHE converts categorical data into binary vectors (numerical formats)
OneHot means only one bit is hot (1) and others are cold (0)

Because machine learning models are mathematical and using numbers to label data makes model biased to some categories like if red = 3 and blue = 1, then ml model will give more priority to red than blue.



Also this increases space complexity of the data set beacause it creates X new columsn, where X is the number of categories.

We only consider X-1 columns beacause absence of other means presence of one

#### Multicollinearity
*Problem:* The variables are highly correlated (perfectly predictable from each other). This breaks linear models (like Linear Regression and Logistic Regression) because they require independent features.

*Solution*: Drop one column (usually the first one). If you have $N$ categories, keep $N-1$ columns.
